## Data Preparation

Machine learning algorithms require data to be numbers, so I'm going to replace 'male' with 0 and 'female' with 1. I'll also generate dummy variables for column 'embarked'.

#### Replacing qualitative data in train set with numbers:

```python
df_train.sex.replace({'male':0, 'female':1}, inplace=True)
df_train = pd.concat([df_train, pd.get_dummies(df_train['embarked'], prefix='embarked')], axis=1)
df_train = df_train.drop(['embarked'], axis=1)
```

```python
df_train.head()
```

#### Doing the same for test set:

```python
df_test.sex.replace({'male':0, 'female':1}, inplace=True)
df_test = pd.concat([df_test, pd.get_dummies(df_test['embarked'], prefix='embarked')], axis=1)
df_test = df_test.drop(['embarked'], axis=1)
```

```python
df_test.head()
```

Now, that all the values are numerical, I can calculate feature correlations with Pearson correlation coefficient.

#### Displaying Pearson Correlation of Features:

```python
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features')
sns.heatmap(df_train.corr(),linewidths=1,
            square=True, cmap='RdBu', linecolor='white', annot=True)
```

#### Specifying independent and dependent variables:

```python
x = df_train.drop('survived', axis=1)
y = df_train['survived']
```

#### Splitting data into training and test set:

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

## Modeling

Let's compare accuracies and metrics to evaluate some machine learning algorithms. I'm going to use sklearn accuracy_score and classification_report for that.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

models = [('LR', LogisticRegression(solver='liblinear')), ('RF', RandomForestClassifier(max_depth=2, random_state=0)),
          ('NB', GaussianNB()), ('KNN', KNeighborsClassifier(n_neighbors=2, algorithm='ball_tree')),
          ('DT', DecisionTreeClassifier(random_state=10)), ('SVC', SVC(gamma='auto'))]

helpers.evaluate(models, x_train, y_train, x_test, y_test)
```

From the data above, we can see that some ML models may be a better fit for our data than the others. We can also observe that train set accuracy is always higher than test set accuracy. Decision Tree Classifier has the highest train set accuracy, almost 100%, but its test set accuracy is significantly lower. This is due to overfitting, i.e. the condition when the model completely fits the training data but fails to generalize the testing unseen data. Therefore, we should choose the model based on test set accuracy.

```python
helpers.compare_on_boxplot(models, x_train, y_train)
```

From the box and whisker plot, we conclude that Logistic Regression and Random Forest have the best mean accuracy and are worthy of further analysis.

## Hyperparameters Tuning

#### GridSearchCV

```python
models = [('LR', LogisticRegression(max_iter=1000)), 
('RF', RandomForestClassifier())]

for name, model in models:
    if name == 'LR':
            solvers = ['newton-cg', 'lbfgs', 'liblinear']
            penalty = ['l2']
            c_values = [100, 10, 1.0, 0.1, 0.01]
            grid = dict(solver=solvers,penalty=penalty,C=c_values)
            lr_parameters = helpers.perform_grid_search_cv(model, grid, x_train, y_train)

    elif name == 'RF':
            n_estimators = [10, 100, 1000]
            max_features = ['sqrt', 'log2']
            max_depth = [3, 5, 10]
            grid = dict(n_estimators=n_estimators, max_features=max_features, max_depth=max_depth)
            rf_parameters = helpers.perform_grid_search_cv(model, grid, x_train, y_train)
            
            
models = [('LR', LogisticRegression(**lr_parameters, random_state=1)), 
('RF', RandomForestClassifier(**rf_parameters, random_state=0))]

helpers.evaluate(models, x_train, y_train, x_test, y_test)
```

After performing hyperparameter tuning using GridSearchCV, Random Forest Classifier looks like the best fit for Titanic Survivor Prediction problem.
